```python
from ultralytics import YOLO

# Load the model.
model = YOLO('yolov8n.pt')

# Training.
results = model.train(
   data='/content/logo_detect_yolov8.yaml',
   epochs=200,
   batch=32,
   name='train',
   lr0=0.001
)
```
